```python
import tweepy
```

```python
api_key = 'XXXX'
api_key_secret = 'xxxx'
access_token = 'xxxx'
access_token_secret = 'xxxx'
```

```python
auth = tweepy.OAuthHandler(api_key, api_key_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth)
```

```python
user = 'ridvancan'
cangoymen_tweets = api.user_timeline(screen_name = user)
print(cangoymen_tweets)
```

```python
query = 'üsküdar'
search_tweets = api.search_tweets(q=query, lang='tr', result_type='popular')
print(search_tweets)
```

```python
query_topic = 'bitcoin'
search_tweets = api.search_30_day(label='development', query=query_topic, fromDate = '202204010000', toDate = '202204100000', maxResults=10)
print(search_tweets)
```

```python
query = 'kampanya'
search_tweets = api.search_tweets(q=query, geocode = '41.021375,29.027874,10km')
print(search_tweets)
```

API Detayları : https://docs.tweepy.org/en/stable/api.html 

Tweetlerin CSV export edilmesi

```python
import pandas as pd

# create dataframe
columns = ['Date', 'User', 'Tweet']

data = []
for tweet in search_tweets:
    data.append([tweet.created_at, tweet.user.screen_name, tweet.text])

df = pd.DataFrame(data, columns=columns)

df.to_csv('tweets.csv')
```
